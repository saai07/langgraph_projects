```python
from typing import TypedDict , List , Dict
from langgraph.graph import StateGraph ,START , END
import random
```

```python
class AgentState(TypedDict):
    name:str
    number:List[int]
    counter:int
    
```

```python
def greeting_node(state :AgentState) ->AgentState:
    """ Greeting node which say hi to the user"""
    state['name'] = f" Hello dear {state['name']} ! , How are you ?"
    state['counter']= 0
    return state

def random_node(state :  AgentState) -> AgentState:
     """Generate the random number"""
     
     state['number'].append(random.randint(0,9))
     state['counter'] += 1
     return state
 
def should_countinue(state : AgentState) -> AgentState:
    
     """ Function decide what to do next"""
     if state['counter'] < 5:
        
        print("Entering loop" , state["counter"])
        return "loop"
     else:
         return "exit"
     
     
```

```python
graph = StateGraph(AgentState)
graph.add_node("greeting", greeting_node)
graph.add_node("random",random_node)
graph.add_edge("greeting" , "random")

graph.add_conditional_edges(
    "random",
    should_countinue,
    {
        "loop" : "random",
        "exit" : END
    }
)

graph.set_entry_point("greeting")

app = graph.compile()
```

```python
result = app.invoke({"name":"vaibhab" , "number":[]})
print(result)
```
